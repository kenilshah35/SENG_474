## Helper functions for plotting errors of various classifiers


```python
import matplotlib.pyplot as plt
import numpy as np

def plot_bar_error(labels, errors, title):
    # plot the train and test errors of labels in a bar graph
    ax = plt.subplot(111)
    _X = np.arange(len(labels))
    colors = ['cornflowerblue','lightgreen']
    for i in range(2):
        ax.bar(_X - 0.25 + i/4.0, errors[i], width=0.25, align='edge', color=colors[i])
    plt.xticks(_X, labels)
    ax.set_ylabel('error')
    ax.legend(['train error', 'test error'])
    ax.set_title(title)
```
